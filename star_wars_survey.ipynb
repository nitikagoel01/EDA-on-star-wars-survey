```python
#importing necessary modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Read the csv file 

star_war = pd.read_csv('/home/nikita/Documents/python projects/StarWars.csv', encoding="ISO-8859-1")
```

```python
#Removing rows where RespondentID is NAN
star_war = star_war[pd.notnull(star_war['RespondentID'])]
star_war.head(5)
```

```python
# Cleaning And Mapping Yes/No Columns

d = {'Yes':'True', 'No':'False'}

star_war['Have you seen any of the 6 films in the Star Wars franchise?'] = star_war['Have you seen any of the 6 films in the Star Wars franchise?'].map(d)
star_war['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_war['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(d)

star_war.head(10)
```

```python
#Cleaning And Mapping Checkbox Columns

column_map = {'Star Wars: Episode I  The Phantom Menace':True,
               'Star Wars: Episode II  Attack of the Clones':True,
               'Star Wars: Episode III  Revenge of the Sith':True,
               'Star Wars: Episode IV  A New Hope':True,
               'Star Wars: Episode V The Empire Strikes Back':True,
               'Star Wars: Episode VI Return of the Jedi':True,
                np.nan : False}

for col in star_war.columns[3:9]:
    star_war[col] = star_war[col].map(column_map)

star_war.head(3)

```

```python
#renaming column

col_name = {'Which of the following Star Wars films have you seen? Please select all that apply.': 'Scene_1',
           'Unnamed: 4' : 'Scene_2',
           'Unnamed: 5' : 'Scene_3',
           'Unnamed: 6' : 'Scene_4',
           'Unnamed: 7' : 'Scene_5',
           'Unnamed: 8' : 'Scene_6'}

star_war.rename(columns = col_name, inplace=True)

```
